```python

class Saldytuvas:
    def __init__(self):
        self.turinys = {}

    def prideti_produktą(self, pavadinimas, kiekis):
        if pavadinimas in self.turinys:
            self.turinys[pavadinimas] += kiekis
        else:
            self.turinys[pavadinimas] = kiekis

    def isimti_produktą(self, pavadinimas, kiekis):
        if pavadinimas in self.turinys:
            self.turinys[pavadinimas] -= kiekis
            if self.turinys[pavadinimas] <= 0:
                del self.turinys[pavadinimas]
        else:
            print("Produktas neegzistuoja šaldytuve.")

    def patikrinti_kieki(self, pavadinimas, norimas_kiekis):
        if pavadinimas in self.turinys and self.turinys[pavadinimas] >= norimas_kiekis:
            print(f"{pavadinimas} yra šaldytuve, turimas kiekis: {self.turinys[pavadinimas]}")
        else:
            print(f"{pavadinimas} nėra šaldytuve arba yra nepakankamai.")

    def spausdinti_turini(self):
        print("Šaldytuvo turinys:")
        for produktas, kiekis in self.turinys.items():
            print(f"{produktas}: {kiekis}")


# Pavyzdinis naudojimas:
saldytuvas = Saldytuvas()
saldytuvas.prideti_produktą("Pienas", 2)
saldytuvas.prideti_produktą("Kiaušiniai", 6)
saldytuvas.prideti_produktą("Obuoliai", 3)

saldytuvas.spausdinti_turini()

saldytuvas.isimti_produktą("Pienas", 1)
saldytuvas.patikrinti_kieki("Pienas", 1)
saldytuvas.patikrinti_kieki("Obuoliai", 5)
```

```python
def prideti_produktą(saldytuvas, produktas, kiekis):
    if produktas in saldytuvas:
        saldytuvas[produktas] += kiekis
    else:
        saldytuvas[produktas] = kiekis

def isimti_produktą(saldytuvas, produktas, kiekis):
    if produktas in saldytuvas:
        if saldytuvas[produktas] >= kiekis:
            saldytuvas[produktas] -= kiekis
            if saldytuvas[produktas] == 0:
                del saldytuvas[produktas]
            return True
        else:
            print("Nepakankamas produktų kiekis šaldytuve.")
            return False
    else:
        print("Produktas nerastas šaldytuve.")
        return False

def patikrinti_kieki(saldytuvas, produktas, kiekis):
    if produktas in saldytuvas and saldytuvas[produktas] >= kiekis:
        print(f"Produktų {produktas} kiekis šaldytuve yra pakankamas.")
        return True
    else:
        print(f"Produktų {produktas} kiekis šaldytuve nepakankamas.")
        return False

def ispausdinti_turini(saldytuvas):
    print("Šaldytuvo turinys:")
    for produktas, kiekis in saldytuvas.items():
        print(f"{produktas}: {kiekis} vnt.")

# Pavyzdys naudojant funkcijas su float kiekiais
saldytuvas = {}

prideti_produktą(saldytuvas, "Obuoliai", 5.5)
prideti_produktą(saldytuvas, "Pienas", 2.3)
prideti_produktą(saldytuvas, "Obuoliai", 3.2)

ispausdinti_turini(saldytuvas)

isimti_produktą(saldytuvas, "Pienas", 1.2)
isimti_produktą(saldytuvas, "Obuoliai", 2.8)

ispausdinti_turini(saldytuvas)

patikrinti_kieki(saldytuvas, "Obuoliai", 1)
patikrinti_kieki(saldytuvas, "Pienas", 3)
```
